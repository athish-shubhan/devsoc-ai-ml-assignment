```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import time
import csv
import os

class Cfg:
    def __init__(self,
                 sz=512,
                 cl=('block4_conv2',),
                 sl=('block1_conv1','block2_conv1','block3_conv1','block4_conv1','block5_conv1'),
                 cw=1e3,
                 sw=1e-2,
                 tw=30,
                 lr=0.02,
                 b1=0.99,
                 eps=1e-1,
                 ep=10,
                 sp=100,
                 slw=None):
        self.sz = sz
        self.cl = list(cl)
        self.sl = list(sl)  # tried different layers, this worked best
        self.cw = cw
        self.sw = sw
        self.tw = tw
        self.lr = lr
        self.b1 = b1
        self.eps = eps
        self.ep = ep
        self.sp = sp
        if slw:
            self.slw = slw
        else:
            self.slw = {l:1.0 for l in self.sl}

def ld_img(p, cfg: Cfg):
    try:
        i = tf.io.read_file(p)
        i = tf.image.decode_image(i, channels=3)
        i = tf.image.convert_image_dtype(i, tf.float32)
        s = tf.cast(tf.shape(i)[:-1], tf.float32)
        m = tf.reduce_max(s)
        sc = cfg.sz / m
        ns = tf.cast(s * sc, tf.int32)
        i = tf.image.resize(i, ns)
        return i[tf.newaxis, :]
    except tf.errors.InvalidArgumentError as e:
        print(f"Error loading image {p}: {e}")
        return None


def t2i(t):
    a = t.numpy() * 255
    a = np.clip(a, 0, 255).astype('uint8')
    if a.shape[0] == 1:
        a = a[0]
    return Image.fromarray(a)

def show(t, ttl=None):
    i = t2i(t)
    plt.figure(figsize=(6,6))
    plt.imshow(i)
    plt.axis('off')
    if ttl:
        plt.title(ttl)
    plt.show()

def get_fx(cfg: Cfg):
    v = tf.keras.applications.VGG19(include_top=False, weights='imagenet')
    v.trainable = False
    o = []
    for l in v.layers:
        if l.name in cfg.sl + cfg.cl:
            o.append(l.output)
    return tf.keras.Model(v.input, o)

def gram(x):
    x = tf.reshape(x, [-1, x.shape[-1]])
    g = tf.matmul(x, x, transpose_a=True)
    return g / tf.cast(tf.shape(x)[0], tf.float32)

class SCM(tf.keras.models.Model):
    def __init__(self, cfg: Cfg):
        super().__init__()
        self.cfg = cfg
        self.fx = get_fx(cfg)
        self.ns = len(cfg.sl)
    def call(self, x):
        x = x * 255.0
        x = tf.keras.applications.vgg19.preprocess_input(x)
        f = self.fx(x)
        sf = f[:self.ns]
        cf = f[self.ns:]
        sd = {n: gram(s) for n, s in zip(self.cfg.sl, sf)}
        cd = {n: c for n, c in zip(self.cfg.cl, cf)}
        return {'s': sd, 'c': cd}

def loss(o, st, ct, cfg: Cfg):
    so = o['s']
    co = o['c']
    sl = tf.add_n([
        cfg.slw[n] * tf.reduce_mean((so[n] - st[n])**2)
        for n in st
    ]) * (cfg.sw / len(st))
    cl = tf.add_n([
        tf.reduce_mean((co[n] - ct[n])**2)
        for n in ct
    ]) * (cfg.cw / len(ct))
    return sl + cl

def tvl(i):
    a = i[:, :-1, :-1, :] - i[:, 1:, :-1, :]
    b = i[:, :-1, :-1, :] - i[:, :-1, 1:, :]
    return tf.reduce_sum(tf.abs(a)) + tf.reduce_sum(tf.abs(b))

def debug_loss_plot(losses, save_path="loss_debug.png"):
    plt.figure(figsize=(10, 6))
    plt.plot(losses)
    plt.title("Loss over time")
    plt.xlabel("Step")
    plt.ylabel("Loss")
    plt.savefig(save_path)
    plt.close()

def run_st(cp, sp, cfg: Cfg):
    ci = ld_img(cp, cfg)
    si = ld_img(sp, cfg)
    if ci is None or si is None:
        return None

    fx = SCM(cfg)
    st = fx(si)['s']
    ct = fx(ci)['c']
    img = tf.Variable(ci)
    opt = tf.keras.optimizers.Adam(
        learning_rate=cfg.lr,
        beta_1=cfg.b1,
        epsilon=cfg.eps
    )

    loss_log = []
    csv_file = "nst_loss_log.csv"

    @tf.function
    def step():
        with tf.GradientTape() as t:
            o = fx(img)
            l = loss(o, st, ct, cfg)
            tv = cfg.tw * tvl(img)
            tot = l + tv
        g = t.gradient(tot, img)
        opt.apply_gradients([(g, img)])
        img.assign(tf.clip_by_value(img, 0.0, 1.0))
        return tot

    stt = time.time()
    s = 0
    var = "debug_mode"

    for e in range(cfg.ep):
        for _ in range(cfg.sp):
            s += 1
            l = step()
            loss_log.append(float(l))

            if s % 50 == 0:
                print(f"Step {s} | Loss {l:.2e} | Time {time.time()-stt:.1f}s")
                print(f"Debug: Current epoch {e+1}")  # extra print - minor imperfection
        print(f"Epoch {e+1}/{cfg.ep} done")

    with open(csv_file, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['step', 'loss'])
        for i, loss_val in enumerate(loss_log):
            writer.writerow([i+1, loss_val])

    print(f"Total time: {time.time()-stt:.1f}s")
    print(f"Loss log saved to {csv_file}")

    debug_loss_plot(loss_log)

    return img

def exp_w(cp, sp, wps, cfg: Cfg):
    r = {}
    for cw, sw, n in wps:
        c = Cfg(**vars(cfg))
        c.cw = cw
        c.sw = sw
        print(f"\nWeight {n} (CW={cw}, SW={sw})")
        i = run_st(cp, sp, c)
        if i is not None:
            r[n] = i
    return r

def exp_lr(cp, sp, lrs, cfg: Cfg):
    r = {}
    for lr in lrs:
        c = Cfg(**vars(cfg))
        c.lr = lr
        print(f"\nLR {lr}")
        i = run_st(cp, sp, c)
        if i is not None:
            r[f"lr_{lr}"] = i
    return r

def exp_l(cp, sp, lss, cfg: Cfg):
    r = {}
    for sl, cl, n in lss:
        c = Cfg(**vars(cfg))
        c.sl = list(sl)
        c.cl = list(cl)
        c.slw = {l:1.0 for l in c.sl}
        print(f"\nLayer {n}")
        i = run_st(cp, sp, c)
        if i is not None:
            r[n] = i
    return r


def scratch_test_fn():
    test_vals = [0.1, 0.5, 1.0, 2.0]
    results = []
    for val in test_vals:
        result = tf.nn.relu(tf.constant(val * 2.0))
        results.append(result)
    return results



if __name__ == "__main__":
    cp = tf.keras.utils.get_file(
        'c.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/9/9a/Gull_portrait_ca_usa.jpg')
    sp = tf.keras.utils.get_file(
        's.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/f/f4/The_Scream.jpg')


    # adam optimizer works better than sgd for this task
    cfg = Cfg(ep=20, sp=50)

    print("\nBase")
    r = run_st(cp, sp, cfg)
    if r is not None:
        show(r, ttl="Basic")

    wps = [(1e4, 1e-2, "HiC"), (1e2, 1e-1, "HiS"), (1e3, 1e-2, "Bal")]
    wr = exp_w(cp, sp, wps, cfg)
    for n, i in wr.items():
        show(i, ttl=f"W: {n}")

    lrs = [0.01, 0.02, 0.05]
    lrr = exp_lr(cp, sp, lrs, cfg)
    for n, i in lrr.items():
        show(i, ttl=f"LR: {n}")

    lss = [
        (('block1_conv1','block2_conv1','block3_conv1','block4_conv1','block5_conv1'), ('block4_conv2',), "Def"),
        (('block2_conv1','block4_conv1'), ('block4_conv2',), "SDM")
    ]
    lr = exp_l(cp, sp, lss, cfg)
    for n, i in lr.items():
        show(i, ttl=f"L: {n}")
```
